```python
from ossapi import *
from scipy import optimize
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import pickle

api = OssapiV2(19003, 'TVMi13ZOzWLvCo1CUMTg4Bs49r0ytI1PZ0INDr0d')
```

```python
countries = np.array([])
users = []
```

```python
for i in api.ranking("osu", RankingType.PERFORMANCE).ranking[0:50]:
    countries = np.append(countries, i.user.country_code)
```

```python
for country in np.unique(countries):
    cursor = None
    for _ in range(20):
        r = api.ranking("osu", RankingType.PERFORMANCE,cursor=cursor,country=country)
        users += r.ranking
        cursor = r.cursor
```

```python
pp = np.zeros(len(users))
time = np.zeros(len(users))
for i in range(len(users)):
    pp[i] = (users[i].pp)
    time[i] = (users[i].play_time)

time /= 3600
```

```python
plt.scatter(time, pp, s = .1)
```

```python
center_pp = pp[(time>np.percentile(time, 10)) & (time<np.percentile(time, 90))]
center_time = time[(time>np.percentile(time, 10)) & (time<np.percentile(time, 90))]
params = optimize.curve_fit(lambda t,a,b: a*t**b, time, pp, p0 = (10,.1))
std = np.sqrt(np.diag(params[1]))
def corr(x, change): 
    return (params[0][0] + change) * x ** params[0][1]
```

```python
x = np.arange(0, 8000)
y_mean = corr(x,0)
plt.plot(x, y_mean, c = 'black')

plt.scatter(center_time, center_pp, s = .1)
```

```python
def find_percentile(change, desired):
    within = 0
    for i in range(len(center_time)):
        if center_pp[i] <= corr(center_time[i], change):
            within += 1

    return abs(within/len(center_time) - desired)

lower_s3 = optimize.minimize_scalar(find_percentile, args = (.0013), tol = 1)
lower_s2 = optimize.minimize_scalar(find_percentile, args = (.0227), tol = 1)
lower_s1 = optimize.minimize_scalar(find_percentile, args = (.1587), tol = 1)
upper_s1 = optimize.minimize_scalar(find_percentile, args = (.8413), tol = 1)
upper_s2 = optimize.minimize_scalar(find_percentile, args = (.9772), tol = 1)
upper_s3 = optimize.minimize_scalar(find_percentile, args = (.9987), tol = 1)
```

```python
a_range = np.array([lower_s3.x, lower_s2.x, lower_s1.x, params[0][0], upper_s1.x, upper_s2.x, upper_s3.x])
```

```python
def stats(user):
    def z_find(z, pp, x):
        return corr(x, z) - pp

    user_pp = api.user(user).statistics.pp
    user_play_time = api.user(user).statistics.play_time/3600
    
    z = optimize.fsolve(z_find, x0 = 0, args = (user_pp, user_play_time))
    print(user_pp, user_play_time, z)
    
    return scipy.stats.norm.cdf(z)
    
```
