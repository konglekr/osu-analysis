```python
countries = np.array([])
users = []
```

```python
for i in api.ranking("osu", RankingType.PERFORMANCE).ranking[0:50]:
    countries = np.append(countries, i.user.country_code)
```

```python
for country in np.unique(countries):
    cursor = None
    for _ in range(20):
        r = api.ranking("osu", RankingType.PERFORMANCE,cursor=cursor,country=country)
        users += r.ranking
        cursor = r.cursor
```

```python
pp = np.zeros(len(users))
time = np.zeros(len(users))
for i in range(len(users)):
    pp[i] = (users[i].pp)
    time[i] = (users[i].play_time)

time /= 3600
```

```python
plt.scatter(time, pp, s = .1)
```

```python
center_pp = pp[(time>np.percentile(time, 10)) & (time<np.percentile(time, 90))]
center_time = time[(time>np.percentile(time, 10)) & (time<np.percentile(time, 90))]
params = optimize.curve_fit(lambda t,a,b: a*t**b, time, pp, p0 = (10,.1))
std = np.sqrt(np.diag(params[1]))
def corr(x, change): 
    return (params[0][0] + change) * x ** params[0][1]
def rev_corr(x, pp):
    return (pp * x ** (-params[0][1])) - params[0][0]
```

```python
def find_percentile(change, desired):
    within = 0
    for i in range(len(center_time)):
        if center_pp[i] <= corr(center_time[i], change):
            within += 1

    return abs(within/len(center_time) - desired)

lower_s2 = optimize.minimize_scalar(find_percentile, args = (.0227), tol = .00001)
lower_s1 = optimize.minimize_scalar(find_percentile, args = (.1587), tol = .00001)
upper_s1 = optimize.minimize_scalar(find_percentile, args = (.8413), tol = .00001)
upper_s2 = optimize.minimize_scalar(find_percentile, args = (.9772), tol = .00001)
```

```python
a_range = np.array([lower_s2.x, lower_s1.x, 0, upper_s1.x, upper_s2.x])
```
